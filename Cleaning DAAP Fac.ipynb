```python
import pandas as pd
import nltk
import sklearn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DAAP_fac.csv')
```

```python
df.head()
```

```python
fac_list = []
for x in range(df.shape[0]):
    fac_list.append(df['Faculty Name'].iloc[x]) 
fac_list
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
stop_words = stopwords.words('english')
import string
rmchars = string.punctuation + "º"
rmchars = rmchars.replace("_", "")
interests_words = {}

def clean_text(doc):
    return doc.lower().translate(str.maketrans("", "", rmchars)).split()


```

```python
rm_words = ['methods', 'practice', 'past', 'daap', 'education', 'mehta', 'taught', 'several', 'arts', 'architecture', 'design', 'area']


def get_unique_words(column, show=False, min_count=3):
    interests = {}
    for x in range(df.shape[0]):
        
        to = clean_text(df[column].iloc[x])
        fac_memb = df["Faculty Name"].iloc[x]
        filtered_sentence = [w for w in to if not w in stop_words] 
        for w in filtered_sentence:
            if w not in rm_words:
                if w in interests:
                    interests[w]['count'] += 1
                    interests[w]['faculty'].append(fac_memb)
                else:
                    interests[w] = { 'count' : 1, 'faculty' : [fac_memb]}
                
    new_interests = {}
    for w in interests:
        if interests[w]['count'] > min_count:
            if len(w) > 3:
                if len(list(set(interests[w]['faculty']))) > 2:
                    new_interests[w] = interests[w]
                    new_interests[w]['faculty'] = list(set(new_interests[w]['faculty']))
        
            
    interests = new_interests
    if show:
        plt.figure(figsize=(100, 5))
        plt.bar(range(len(interests)), list(interests.values()), align='center')
        plt.xticks(range(len((interests)), list((interests.keys()), rotation=70)))
        plt.show()
   
    return interests
                         
def expand_cleaned_dict(the_dict):
    l = []
    for i in range(0, len(fac_list)):
        l.append([])

    for x in the_dict:
        for fac in the_dict[x]['faculty']:
            fac_index = fac_list.index(fac)
            l[fac_index].append(x)

    return l
            
```

```python
fac_int = get_unique_words('Interests (5 to 10)', min_count=3)
fac_skills = get_unique_words('Skills', min_count=3)
fac_projects = get_unique_words('Projects', min_count=5)
fac_bios = get_unique_words('Bio', min_count=5)

#for x in fac_bios:
#    print(x)
#    print(fac_bios[x]['count'])
print(len(fac_int))
print(len(fac_skills))
print(len(fac_projects))
print(len(fac_bios))


the_dicts = [fac_int, fac_skills, fac_projects, fac_bios]
the_names = ["interests_cleaned", "skills_cleaned", "projects_cleaned", "bios_cleaned"]
```

```python
the_col = expand_cleaned_dict(fac_int)

for i in range(len(the_dicts)):
    new_col = expand_cleaned_dict(the_dicts[i])
    new_label = the_names[i]

    df[new_label] = new_col

```
