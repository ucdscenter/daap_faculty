```python
import pandas as pd
import nltk
import sklearn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DAAP_fac.csv')
```

```python
df.head()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
stop_words = stopwords.words('english')
import string
rmchars = string.punctuation + "º"
interests_words = {}

def clean_text(doc):
    return doc.lower().translate(str.maketrans("", "", rmchars)).split()


```

```python
def get_unique_words(column):
    interests = {}
    for x in range(df.shape[0]):
        print(df[column].iloc[x])
        to = clean_text(df[column].iloc[x])
        filtered_sentence = [w for w in to if not w in stop_words] 
        for w in filtered_sentence:
            if w in interests:
                interests[w] += 1
            else:
                interests[w] = 1
                
    for w in interests:
        del interests[w]
    return interests
```

```python
fac_int = get_unique_words('Projects')
plt.figure(figsize=(100, 5))
plt.bar(range(len(fac_int)), list(fac_int.values()), align='center')
plt.xticks(range(len(fac_int)), list(fac_int.keys()), rotation=70)
# # for python 2.x:
# plt.bar(range(len(D)), D.values(), align='center')  # python 2.x
# plt.xticks(range(len(D)), D.keys())  # in python 2.x

plt.show()
```
